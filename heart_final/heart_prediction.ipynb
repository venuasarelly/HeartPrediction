```python
#now we need to convert original data into dummy variables
dataset2 = pd.get_dummies(dataset , columns = ['sex' , 'cp' , 'fbs' ,'restecg','exang', 'slope', 'ca', 'thal'])
```

```python
dataset2.head()
```

```python
pd.set_option('display.max_columns', None)
```

```python
display(dataset2)
```

```python
#feature scaling
dataset2.head()
```

```python
cate_val = []
cont_val = []

for column in dataset.columns:
    if dataset[column].nunique() <=10:
        cate_val.append(column)
    else:
        cont_val.append(column)
```

```python
from sklearn.preprocessing import StandardScaler
st = StandardScaler()
dataset2[cont_val] = st.fit_transform(dataset2[cont_val])
```

```python
dataset2.head()
```

```python
#splitting the dataset into the training set and test set
x = dataset2.drop('target',axis=1)
y = dataset2['target']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2 ,random_state=42)
```

```python
x_test
```

```python
x_train
```

```python
#logistic regression
from sklearn.linear_model import LogisticRegression
```

```python
logistic = LogisticRegression()
logistic.fit(x_train,y_train)
```

```python
y_pred1 = logistic.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred1)
```
